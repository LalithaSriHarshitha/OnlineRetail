```python
import pandas as pd
df = pd.read_csv("OnlineRetail.csv", encoding='ISO-8859-1')
df['Description'].fillna('Unknown') 
df.dropna(subset=['CustomerID'], inplace=True)    
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce', dayfirst=True)
df['CustomerID'] = df['CustomerID'].astype(str)        
df['Description'] = df['Description'].str.lower()       
df['Description'] = df['Description'].str.strip()     
df['Description'] = df['Description'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True) 
df.drop_duplicates(inplace=True)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df.to_csv('cleaned_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('cleaned_data.csv')
plt.figure(figsize=(12, 6))
country_counts = df['Country'].value_counts().sort_values(ascending=False)
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10], color="skyblue",legend=False)
plt.title("Top 10 Countries by Number of Costumers")
plt.xlabel("Country")
plt.ylabel("Number of Costumers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('cleaned_data.csv')
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='UnitPrice', data=df[df['Country'].isin(df['Country'].value_counts().head(10).index)])
plt.title('Finding Outliers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
